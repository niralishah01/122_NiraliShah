```python
#Declare label as last column in the source file
label=data.iloc[:,-1].values
print(label)
```

```python
#Declaring X as all columns excluding last
X=data.iloc[:,:-1].values
print(X)
```

```python
# Splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,label,test_size=0.30,random_state=122)
```

```python
X_train,X_test,Y_train,Y_test
```

```python
# Sacaling data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Variabes to calculate sigmoid function
y_pred = []
len_x = len(X_train[0])
w = []
b = 0.2
print(len_x)
```

```python
entries = len(X_train[:,0])
entries
```

```python
for weights in range(len_x):
    w.append(0)
w
```

```python
# Sigmoid function
def sigmoid(z):
  res = (1/(1+np.exp(-z)))
  return res 
def predict(inputs):
  z=sigmoid(np.dot(inputs,w)+b)
  return z
```

```python
#Loss function
def loss_func(y,a):
    J = -(y*np.log(a) + (1-y)*np.log(1-a))
    return J
dw = []
db = 0
J = 0
alpha = 0.1
for x in range(len_x):
    dw.append(0) 
```

```python
#Repeating the process 3000 times
for iterations in range(3000):
    for i in range(entries):
        tmp = X_train[i]
        a = predict(tmp)   
        dz = a - Y_train[i]
        J += loss_func(Y_train[i],a)
        for j in range(len_x):
            dw[j] = dw[j]+(tmp[j]*dz)
        db += dz
    J = J/entries
    db = db/entries
    for x in range(len_x):
        dw[x]=dw[x]/entries
    for x in range(len_x):
        w[x] = w[x]-(alpha*dw[x])
    b = b-(alpha*db)         
    J=0
```

```python
#Print weight
print(w)
#print bias
print(b)
```

```python
#predicting the label
for x in range(len(y_test)):
    y_pred.append(predict(X_test[x]))
```

```python
#print actual and predicted values in a table
print('Actual \t Predicted')
for x in range(len(y_pred)):
  print(y_pred[x],'\t',label[x])
  if y_pred[x]>=0.5:
    y_pred[x]=1
  else:
    y_pred[x]=0
```

```python
# Calculating accuracy of prediction
# count = 0
# for x in range(len(y_pred)):
#     if(y_pred[x]==y_test[x]):
#         count=count+1
# print('Accuracy:',(count/(len(y_pred)))*100)
diff = y_pred - y_test
acc=1.0 - (float(np.count_nonzero(diff)) / len(diff))
print("Accuracy: ",acc)
```
